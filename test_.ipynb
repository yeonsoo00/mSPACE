```python
def getcolorlist():
    # Define colormaps
    cmap1 = plt.cm.tab20(np.linspace(0, 1, 10))
    cmap2 = plt.cm.tab20b(np.linspace(0, 1, 10))
    cmap3 = plt.cm.tab20c(np.linspace(0, 1, 10))

    # Combine them into one array of colors
    all_colors = np.vstack([cmap1, cmap2, cmap3])
    hex_colors = [f"#{int(255 * r):02x}{int(255 * g):02x}{int(255 * b):02x}" for r, g, b, a in all_colors]

    return hex_colors
def buildgraphfromprocessfile(locationlist,graph_node_id_name, celltype,colored_node_ids):
    colorlist = getcolorlist()
    G=nx.DiGraph()
    for nodeid, (nodename, loc) in tqdm(enumerate(zip(graph_node_id_name, locationlist)), total=len(locationlist), desc="Creating Di-graph"):
        G.add_node(nodeid,pos=loc,color=colorlist[celltype[nodename]]if nodeid in colored_node_ids else"#D3D3D3")

    return G
```

```python
def plot_directed_graph(ligand, receptor):
    import pandas as pd
    import networkx as nx
    import matplotlib.pyplot as plt

    # Load datasets
    df_comm = pd.read_csv('/home/yec23006/projects/research/app/mSpace/CLARA/Dataset/Testdata/Mapped_new/csv/Mapped_new_0.08_CCI_details.csv')  # Communication data
    df_cells = pd.read_csv('/home/yec23006/projects/research/app/mSpace/assets/data_09012023/geneexpression.csv')  # Cell coordinates

    # Filter by ligand and receptor
    filtered_comm = df_comm[(df_comm['Ligand'] == ligand) & (df_comm['Receptor'] == receptor)]

    # Create a directed graph
    G = nx.DiGraph()

    # Add all cells as nodes
    for _, row in df_cells.iterrows():
        G.add_node(row['cellid'], pos=(row['cellx'], row['celly']))

    # Add edges based on communication data and mark communicating cells
    for _, row in filtered_comm.iterrows():
        G.add_edge(row['Sender'], row['Receiver'], weight=row['CommunicationScore'])

    # Node colors - add alpha to gray nodes
    node_colors = [(0.585, 0.645, 0.65, 0.5) if node not in set(filtered_comm['Sender']) | set(filtered_comm['Receiver']) else (0.203, 0.596, 0.858, 1) for node in G.nodes()]

    # Position of nodes
    pos = {node: (data['pos'][0], data['pos'][1]) for node, data in G.nodes(data=True)}

    # Draw the graph
    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, node_color=node_colors, node_size=50, 
            edge_color='black', linewidths=1, 
            arrows=True, arrowsize=10, with_labels=False)
    plt.title('Directed Graph of Communicated Cells')
    plt.show()

# Example usage
plot_directed_graph('EFNB1', 'EPHB2')
```
